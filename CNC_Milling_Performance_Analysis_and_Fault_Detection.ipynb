```python
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Load the dataset
data = pd.read_csv('final_data.csv')

# Drop unnecessary columns safely
data = data.drop(columns=['Unnamed: 0'], errors='ignore')

# Label Encoding for categorical variables
label_encoders = {}
categorical_columns = ['tool_condition', 'machining_finalized', 'passed_visual_inspection']

for column in categorical_columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

    # Save each label encoder separately
    with open(f'label_encoder_{column}.pkl', 'wb') as file:
        pickle.dump(le, file)

print("Label encoders saved successfully!")

# Identify and drop non-numeric columns (if necessary)
non_numeric_columns = data.select_dtypes(include=['object']).columns
data = data.drop(columns=non_numeric_columns)

# Separate features and labels
X = data.drop(columns=categorical_columns)
y = data[categorical_columns]

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Save the scaler
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)

print("Scaler saved successfully!")

# Reshape data for LSTM (samples, timesteps, features)
X_reshaped = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42, stratify=y)

print("Data preprocessing completed successfully!")
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, concatenate

# Define the input shape
input_shape = (X_train.shape[1], X_train.shape[2])

# Define the input layer
inputs = Input(shape=input_shape)

# Add LSTM layers
lstm1 = LSTM(50, return_sequences=True)(inputs)
dropout1 = Dropout(0.2)(lstm1)
lstm2 = LSTM(50, return_sequences=False)(dropout1)
dropout2 = Dropout(0.2)(lstm2)

# Add Dense layers for multi-output classification
dense = Dense(50, activation='relu')(dropout2)

# Define separate output layers for each target
output1 = Dense(len(label_encoders['tool_condition'].classes_), activation='softmax', name='tool_condition')(dense)
output2 = Dense(len(label_encoders['machining_finalized'].classes_), activation='softmax', name='machining_finalized')(dense)
output3 = Dense(len(label_encoders['passed_visual_inspection'].classes_), activation='softmax', name='passed_visual_inspection')(dense)

# Define the model
model = Model(inputs=inputs, outputs=[output1, output2, output3])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss={
                  'tool_condition': 'sparse_categorical_crossentropy',
                  'machining_finalized': 'sparse_categorical_crossentropy',
                  'passed_visual_inspection': 'sparse_categorical_crossentropy'
              },
              metrics={
                  'tool_condition': ['accuracy'],
                  'machining_finalized': ['accuracy'],
                  'passed_visual_inspection': ['accuracy']
              })

# Split y_train and y_test into separate arrays for each target
y_train_tool = y_train['tool_condition']
y_train_machining = y_train['machining_finalized']
y_train_inspection = y_train['passed_visual_inspection']

y_test_tool = y_test['tool_condition']
y_test_machining = y_test['machining_finalized']
y_test_inspection = y_test['passed_visual_inspection']

```

```python
# Train the model
history = model.fit(X_train,
                    {'tool_condition': y_train_tool,
                     'machining_finalized': y_train_machining,
                     'passed_visual_inspection': y_train_inspection},
                    epochs=20,
                    batch_size=32,
                    validation_data=(X_test,
                                    {'tool_condition': y_test_tool,
                                     'machining_finalized': y_test_machining,
                                     'passed_visual_inspection': y_test_inspection}),
                    verbose=1)
```
